#### Q1. Explain GET and POST methods.

GET and POST are two of the most commonly used HTTP request methods used in web development. They differ in the way they transmit data to the server and in their intended use cases.

 * GET Method:
GET requests are used to retrieve information from a server. When a GET request is made, the browser sends a request to the server for a specific resource or information. The server then responds to the browser with the requested information. GET requests are considered "safe" requests as they do not modify the state of the resource on the server.
The data is transmitted via the URL in a GET request, which means that any data that is sent is visible in the URL. This makes GET requests less secure for sending sensitive data, such as passwords or credit card information.


 * POST Method:
POST requests are used to submit data to the server. When a POST request is made, the data is sent in the request body, which is not visible in the URL. This makes POST requests more secure for sending sensitive data.
POST requests are commonly used for submitting forms, creating new resources on the server, or updating an existing resource. Unlike GET requests, POST requests can modify the state of the resource on the server.

In summary, GET is used to retrieve data from the server, and POST is used to submit data to the server. GET requests are used when no state modification is required, while POST requests are used when creating, modifying or deleting resources on the server.

#### Q2. Why is request used in Flask ?

In Flask, requests are used to handle client requests sent to the server. A request is an HTTP request made by a client to a server, requesting information or an action. Flask provides a request object that represents the client request and allows you to access information about the request, such as the URL, headers, form data, and query parameters.

#### Q3. Why is redirect() used in Flask ?

In Flask, the redirect() function is used to redirect the client to a different URL or route. When the redirect() function is called, the client's browser is instructed to make a new request to the URL or route specified in the redirect() function. This is useful when you want to send the user to a different page or resource after they have completed a specific action, such as submitting a form or logging in.

#### Q4. What are templates in Flask? Why is the render_template() function used ?

In Flask, templates are used to generate dynamic HTML content and provide a way to separate the presentation layer from the application logic. A template is a file that contains HTML markup and placeholders for dynamic content. When the Flask application renders a template, it replaces the placeholders with the actual data before sending the resulting HTML to the client.

The Flask render_template() function is used to render a template and return the resulting HTML as a response to the client. This function takes the name of the template file and any context data that should be passed to the template. The context data can include variables, lists, and dictionaries that are used to populate the placeholders in the template.
