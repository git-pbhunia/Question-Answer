#### Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is the process of mapping URLs to functions in a Flask application. When a user requests a specific URL, Flask uses the app routing system to determine which function should handle the request and generate the appropriate response.

App routing is a critical aspect of building Flask applications because it determines how users interact with the application. Flask provides a simple and flexible way to define routes using the @app.route decorator. This decorator specifies the URL pattern and the function that should handle requests to that URL pattern.

```python
#Example

from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'This is the home page.'

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    app.run()
```

In this example, there are two routes defined using the @app.route decorator. The first route maps the root URL / to the index() function, and the second route maps the URL /about to the about() function. When a user visits the root URL, the index() function is called, and when a user visits the /about URL, the about() function is called.

We use app routes in Flask to create a mapping between URLs and the corresponding function to handle requests. This allows us to create dynamic and interactive web applications that respond to user input and provide different content depending on the URL requested. By using app routing, we can make our Flask applications more modular and easier to maintain, as different parts of the application can be handled by separate functions or modules.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function in Flask is used for URL building. It generates a URL based on the name of a function and the arguments provided to it. It takes the name of a view function as its first argument, followed by keyword arguments representing the variables to be inserted into the URL.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return f'The URL for the about page is: {url_for("about")}'

@app.route('/about')
def about():
    return 'This is the about page.'

if __name__ == '__main__':
    app.run()
```
