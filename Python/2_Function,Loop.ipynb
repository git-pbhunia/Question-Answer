### Solution-1 :

```python
#A program to display the grade given the marks

#Function to calculate the grade according to the criteria
def grade(marks):
    if marks>90:
        print("Grade: A")
    elif marks>80 and marks<=90:
        print("Grade: B")
    elif marks>=60 and marks<=80:
        print("Grade: C")
    else:
        print("Grade: D")
    return

#Taking the user input
marks = int(input("Enter the marks: "))

#Output
grade(marks)
```

### Solution-2 :

```python
#A program to display the road tax given the cost of a bike

#Function to calculate the road tax according to the criteria
def road_tax(cost):
    if cost>100000:
        tax = cost*0.15
    elif cost>50000 and cost<=100000:
        tax = cost*0.10
    else:
        tax = cost*0.05
    return print("Tax payable(Rs): ", tax)

#Taking the user input
cost = float(input("Enter the cost of the bike(Rs): "))

#Output
road_tax(cost)
```

```python
#A program to display the monument given the city

#Function to display the monument according to the city
def city_monument(city):
    if city=="Delhi":
        print("Monument: Red Fort")
    if city=="Agra":
        print("Monument: Taj Mahal")
    if city=="Jaipur":
        print("Monument: Jal Mahal")
    return

#Taking the user input
city = input("Enter the name of the city (Delhi/Agra/Jaipur): ")

#Output
city_monument(city)
```

### Solution-4 :

```python
#A program to check how many times a given number can be divided by 3 before it is less than or equal to 10

def check_divisible_by_3(num):
    count = 0
    while num > 10:
        num = num / 3
        count += 1
    return count

num = int(input("Enter a number: "))
result = check_divisible_by_3(num)

print("The number of time(s) the number can be divided by 3 before it is less than or equal to 10: ", result)
```

### Solution-5 :

##### The while loop in Python is used when you want to repeat a block of code multiple times until a certain condition is met. In other words, you can use a while loop when you don't know beforehand how many times you need to repeat the code, and you want to repeat it as long as a certain condition is true.


```python
#Example: Repeat until a number is divisible by 7

number = int(input("Enter a number: "))
while number % 7 != 0:
    number = int(input("Enter a number: "))
    
print("The number you entered is divisible by 7.")
```

### Solution-6 :

```python
#Pattern 1: Right Triangle

rows = int(input("Enter the number of rows: "))
i = 1
while i <= rows:
    j = 1
    while j <= i:
        print("*", end="")
        j += 1
    print()
    i += 1
    
```

```python
#Pattern 2: Inverted Right Triangle

rows = int(input("Enter the number of rows: "))
i = rows
while i >= 1:
    j = i
    while j >= 1:
        print("*", end="")
        j -= 1
    print()
    i -= 1
    
```

```python
#Pattern 3: Diamond

rows = int(input("Enter the number of rows: "))
k = 0
for i in range(1, rows * 2):
    for j in range(rows - abs(rows - i)):
        print(" ", end="")
    while k != (2 * (rows - abs(rows - i)) - 1):
        print("*", end="")
        k += 1
    k = 0
    print()
```
